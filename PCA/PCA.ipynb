# **Introduction**
Principal Component Analysis (PCA) is a statistical technique used for dimensionality reduction while preserving the maximum variance in the data. It is widely used in various domains like image processing, finance, genetics, and speech recognition. This document provides a detailed explanation of PCA, its applications, working principles, and its role in reducing computational complexity by selecting important features from a dataset.
# Example of PCA Implementation
# PCA on a Simple Dataset:

Let's say we have a dataset with 15 features. After applying **Singular Value Decomposition (SVD)** on the dataset, we select 6 important components. By visualizing the results, PCA makes it easier to identify patterns and simplifies the dataset for better analysis.
Data Standardization:

Before applying PCA, it is essential to standardize the dataset. Without transformation, PCA would not work effectively, as it is sensitive to the scale of the data.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
data=pd.read_csv('/content/Wine.csv')
data.head()
```

```python
data.isnull().sum()
```

```python
X=data.iloc[:,:-1]
y=data.iloc[:,-1]
target = data["Customer_Segment"]   # labels (1,2,3)
target_names = ["Class 1", "Class 2", "Class 3"]
```

```python
scaler=StandardScaler()
scaled_data=scaler.fit_transform(X)
```

```python
pca=PCA(n_components=2)
X_pca=pca.fit_transform(scaled_data)
```

```python
plt.figure(figsize=(8, 6))

colors = ['navy', 'turquoise', 'darkorange']

for color, i, target_name in zip(colors, [1, 2,3], target_names):

    plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1], alpha=0.7, color=color, label=target_name)



plt.legend()

plt.title('PCA of Wine Dataset')

plt.xlabel('Principal Component 1')

plt.ylabel('Principal Component 2')

plt.grid(True)

plt.tight_layout()

plt.show()

```

```python
print("Explained variance ratio:", pca.explained_variance_ratio_)
```
