DATA PREPROCESSING

STEP 1
Data loading

```python
import pandas as pd
```

```python
df=pd.read_csv("/content/Students Social Media Addiction.csv")
df.head()
```

```python
df.dtypes
```

#STEP 2
 Removing duplicate rows

```python

df_cleaned=df.drop_duplicates()
df_cleaned.head()
```

STEP 3
Handling Missing Vlaues

```python
df_cleaned.isnull().sum()
#currently this data has no missing value
```

```python
#In case missing values the commands are:
df_cleaned['Age'].fillna(df_cleaned['Age'].mean(), inplace=True)
df_cleaned['Country'].fillna(df_cleaned['Country'].mode(), inplace=True)

#Now check again for missing values
df_cleaned.isnull().sum()
```

STEP 4
Normalization (0-1)
Standardization(mean=0 , std=1)

```python
#Normalization
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df_cleaned[['Age']]=scaler.fit_transform(df_cleaned[['Age']])
df_cleaned.head()
```

```python
#Sacling
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
df_cleaned[['Age']]=scaler.fit_transform(df_cleaned[['Age']])
df_cleaned[['Age']].head()
```

STEP 5
Encoding Categorical Variables

Label Encoding: Convert categorical variables into numeric form (Ordinal) e.g: Gender 0,1 or days of the week mod,tues,wed,thurs = 0,1,2,3
One Hot Encoding: Converting each category as separate binary feature e.g: create 7 cols for days of the week and put 0 or 1 where days is present or not

```python
#Label Encoding
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
df_cleaned['Gender']=encoder.fit_transform(df_cleaned['Gender'])
df_cleaned.head()
```

```python
#One Hot Encoding
import pandas as pd
df_cleaned = pd.get_dummies(df_cleaned, columns=['Country'])

# View result
df_cleaned.head()
```
